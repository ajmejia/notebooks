##Algoritmo

El procedimiento es bastante sencillo:
```
para cada espectro J-PAS (lambda, flux):
    para cada i=1,...,nlambda - 2:
        K[i] = 1 - (flux[i]/flux[i+1] + flux[i+2]/flux[i])*0.5
        
    donde K > 0.04:
        enmascarar flux
```

##Una medida del éxito
