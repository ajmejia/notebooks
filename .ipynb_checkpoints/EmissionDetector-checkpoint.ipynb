#Emission line detector
##Introduction
This is a code to test recoverability of the emission line information, namely, equivalent widths, intensities, etc...
The idea is rather simple: since the emission lines in J-PAS spetra appears as an excess in three passbands, and this three bands form a triangle with a particular shape, it should be easy to parametrize it as a function of two parameters $k_1$ and $k_2$.

##Approaches
###Trigonometric parameters
As a first approach this parameters were defined to be
$$\beta\equiv k_1-k_2,$$
where $\beta$ is the angle as shown in the figure below.

###Color parameters
This time, the parameters were defined to be
$$k_1\equiv f_1/f_2\text{ and }k_2\equiv f_3/f_1,$$

###Jesús' suggestion
During a Student Meeting, was suggested the computation of the equivalent width from the J-PAS spectra...

```python
import os
import pyfits as pyf
from scipy.interpolate import interp1d as linear
from scipy.optimize import fmin as minimum
from scipy.optimize import leastsq

JUMP_SIZE = 5
N_PARS    = 2
N_STEPS   = 10000
SKIP      = 10

np.seterr(all="ignore")

def mAB(flux, wlength):
    return 22.5 - 2.5*log10(flux * (wlength**2 / 3e18) * 1e23 / 3.631e-6)

def rejection(pdf, domain=(0., 1.)):
    """Compute random deviate from pdf using rejection technique"""
    max_pdf = pdf(minimum(lambda x: -pdf(x), 0.5*(domain[1]+domain[0]), disp=False))[0]
    
    u = random.rand()
    x = domain[0] + random.rand()*(domain[1] - domain[0])
    while u > pdf([x])[0] / max_pdf:
        u = random.rand()
        x = domain[0] + random.rand()*(domain[1] - domain[0])
    return x

def eval_poly(x, coeff):
    coeff = coeff[::-1]
    polyn = coeff[0]
    for i in xrange(1, coeff.size): polyn += coeff[i]*x**i
    return polyn
```

```python
def draw_from_prior():
    k1 = rejection(pdf1, domain=(0.6, 1.0))
    k2 = rejection(pdf2, domain=(0.7, 1.0))

    return np.array([k1, k2])

def log_prior(params):
    k1, k2 = params[0], params[1]
    
    return pdf1([k1])[0]*pdf2([k2])[0]

def log_likelihood(data, params):
    k1, k2 = params[0], params[1]

    mu    = zeros(3)
    mu[0] = data[0, 1]
    mu[1] = data[0, 1] / k1
    mu[2] = k2 * data[0, 1]
    
    return -0.5*data.shape[0]*log(2.*pi) - sum(log(data[:, 2])) - 0.5*sum((data[:, 1] - mu)**2/data[:, 2]**2)

def proposal(params):
    new = copy(params)

    which = random.randint(N_PARS)
    new[which] += JUMP_SIZE*10.**(1.5 - 6.*random.rand())*random.randn()

    return new
```

```python
data = loadtxt(sdir+"processed_JPAS_spSpec-51695-0351-131.txt", usecols=(0, 1, 2))

wj = data[:, 0]
fj = data[:, 1]

#data[:, 0] /= data[27, 0]
#data[:, 1] /= data[27, 1]
#data[:, 2] /= data[27, 2]

data = data[26:29, :]

params = draw_from_prior()
print "first step: ", params
logp, logl = log_prior(params), log_likelihood(data, params)
print "prior, likelihood: ", logp, logl

sample_params = []
for i in xrange(N_STEPS):
    # Generate proposal
    new = proposal(params)

    # Evaluate prior and likelihood for the proposal
    logp_new, logl_new = log_prior(new), log_likelihood(data, new)

    # Acceptance probability
    log_alpha = (logl_new - logl) + (logp_new - logp)
    if log_alpha > 0.: log_alpha = 0.

    # Accept?
    if random.rand() <= exp(log_alpha):
        params = new
        logp   = logp_new
        logl   = logl_new

    # Save results
    if (i+1)%SKIP == 0:
        sample_params.append(list(params) + [logl])
        #print params, logl

sample_params = array(sample_params)
```
