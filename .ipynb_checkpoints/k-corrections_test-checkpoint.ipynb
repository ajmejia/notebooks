```python
u_obs, g_obs, r_obs, i_obs, z_obs, redshift = loadtxt("../../degree_thesis/data/photometry/total_photometry.txt", usecols=range(1, 6)+[11], unpack=True)
u_obf, g_obf, r_obf, i_obf, z_obf, rs_sam = loadtxt("../outputs/real_deal6/ugriz_photo_of.txt", usecols=range(1, 6)+[11], unpack=True)
u_ref, g_ref, r_ref, i_ref, z_ref = loadtxt("../outputs/real_deal6/ugriz_photo_rf.txt", usecols=range(1, 6), unpack=True)
```

```python
zmask = redshift < 0.03
ocolor_x = (u_obs - g_obs)[zmask]
ocolor_y = (g_obs - r_obs)[zmask]
icolor_x = u_obf - g_obf
icolor_y = g_obf - r_obf
kcolor_x = u_ref - g_ref
kcolor_y = g_ref - r_ref
```

```python
fig = figure(figsize=(7, 7))
xlm = array((0.6, 2.3))
ylm = array((0.1, 1.4))
axs = subplot(111, xlim=xlm, ylim=ylm)
ins = inset_axes(axs, width="2.5%", height="40%", loc=2)

axs.set_xlabel(r"u-g")
axs.set_ylabel(r"g-r")

nx, xi, xf, bsx = nbins(ocolor_x)
ny, yi, yf, bsy = nbins(ocolor_y)

xbin = linspace(xi, xf, 51)
ybin = linspace(yi, yf, 51)

#X, Y = meshgrid(xbin, ybin)
#R = griddata(icolor_x, icolor_y, kcolor_x-icolor_x + kcolor_y-icolor_y, xbin, ybin)

H, xe, ye = histogram2d(ocolor_x, ocolor_y, bins=(nx, ny), normed=True)
H = gaussian_filter(H, sigma=2)

cont = axs.contourf(xe[:- 1], ye[:- 1], H.T, cmap=cm.gray_r, levels=np.arange(1, 11, 1))
cb = colorbar(cont, cax=ins)
cb.set_label(r"galaxies per bin", size="small")

mask = rs_sam < 0.1
#axs.quiver(icolor_x[mask], icolor_y[mask], (kcolor_x-icolor_x)[mask], (kcolor_y-icolor_y)[mask], scale_units="xy", angles="xy", scale=1, headwidth=2)
#iX, iY = meshgrid(icolor_x[mask], icolor_y[mask])
#kX, kY = meshgrid(kcolor_x[mask], kcolor_y[mask])

#U = kX - iX
#V = kY - iY
#axs.streamplot(X, Y, U, V)
axs.scatter(icolor_x[mask], icolor_y[mask], lw=0, c="magenta", s=1)
axs.scatter(kcolor_x[mask], kcolor_y[mask], lw=0, c="chartreuse", s=1)
#axs.plot(xlm, 2.22 - xlm, "--", color="#D65C5C")

ins.tick_params(right=True, left=False, direction="out", labelright="on", labelleft="off")
ins.yaxis.set_label_position("right")

savefig("kcorr_eval.png", bbox_inches="tight")
```
